# loading data using tensorflow_datasets

```python
import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import numpy as np
```

```python
Image_size = 256
Batch_size = 32
channels = 3
epochs = 50
```

```python
## loading data using tenforflow dataset api and resizing all images in 256,256 with batch size = 32
tf_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "PlantVillage",
    shuffle = True,
    image_size = (Image_size,Image_size),
    batch_size = Batch_size
)
```

1. observation 2152 files so furthet to get more data we will do data augmentation.
2. 3 classes of potato are present - Potato__Early_blight , potato__Late_blight , potato_healthy

```python
class_names = tf_dataset.class_names
```

# splitting dataset in three parts (train-validation-test split)
1. training dataset = 80% of original data
2. validation dataset = 10% of original data
3. testing dataset = 10% of original data

```python
#function to split dataset 
def train_val_testsplit(dataset, train = 0.8 , val = 0.1 , test = 0.1 , shuffle = True, shuffle_size=10000):
  dataset_size = len(dataset) #total size of dataset
  if shuffle:
    dataset = dataset.shuffle(shuffle_size,seed = 12)

  train_size = int(dataset_size * train) # size of training dataset
  val_size = int(dataset_size * val)
  test_size = int(dataset_size * test)
  train_ds = dataset.take(train_size)
  val_ds = dataset.skip(train_size).take(val_size) # work as arr[54:] , skipping train_ds 
  test_ds = dataset.skip(train_size).skip(val_size)
  return train_ds, val_ds , test_ds
```

```python
train_ds , val_ds , test_ds = train_val_testsplit(tf_dataset)
```

```python
#cache -> to avoid same images again from memory
#prefetc -> to read and train data simultaneoulsy in cpu and gpu
# it will overall optimise the process of training and reading images from dataset
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

# PreProcessing 

### Resizing and Rescaling pipeline

```python
resize_and_rescale = tf.keras.Sequential([
                  layers.experimental.preprocessing.Resizing(Image_size,Image_size), #to take care of different size images after production
                  layers.experimental.preprocessing.Rescaling(1.0/255.0) # rescaling images to 0 to 1
                                          
])
```

### Data augmentation

```python
# to make our model more robust so it can work on any type of image i.e rotated, fliped etc.
data_augmentation = tf.keras.Sequential([
                                         layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
                                         layers.experimental.preprocessing.RandomRotation(0.2)
])
```

# Building model using CNN architecture

1. 6 CNN layers , 6 pooling Layers , one dense layer  ,and one ouput layer.
2. activation function = relu , for ouput layer = 'softmax'
3. number of classes = 3
4. input_shape = 32,256,256,3

```python
input_shape = (Batch_size , Image_size , Image_size , channels)
n_class = 3
model = models.Sequential([
    resize_and_rescale,

    data_augmentation,

    layers.Conv2D(32,(3,3), activation='relu',input_shape = input_shape),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(n_class , activation= "softmax")

                             
])

model.build(input_shape=input_shape)
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    metrics = ['accuracy']
)
```

```python
history  = model.fit(
    train_ds,
    batch_size = Batch_size,
    epochs = epochs,
    verbose = 1,
    validation_data = val_ds
)
```

```python
scores = model.evaluate(test_ds)
```
